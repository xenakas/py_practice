```python
# Create lists first and second
first = [11.25, 18.0, 20.0]
second = [10.75, 9.50]

# Paste together first and second: full
full=first+second

# Sort full in descending order: full_sorted
full_sorted=sorted(full,reverse=True)

# Print out full_sorted
print(full_sorted)
```

```python
# string to experiment with: room
room = "poolhouse"

# Print out room and room_up
print(room.upper())

# Print out the number of o's in room
print(room.count("o"))
```

```python
# Create list areas
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

areas.append(14.5) #adds an element to the list it is called on
areas.reverse() #reverses the order of the elements in the list it is called on
areas.remove(18.0) #removes the first element of a list that matches the input

print(areas)

# Print out the index of the element 20.0
print(areas.index(20.0))

# Print out how often 14.5 appears in areas
print(areas.count(14.5))
```

## Plots

```python
# generate random integer values
import numpy.random as npr

# seed random number generator
npr.seed(1)

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

# generate some integers
life_exp =  npr.gumbel(45, 20, 100)
gdp_cap = npr.normal(9000, 2000, 100)
pop = npr.randint(0, 20, 100)
col = npr.choice(colors, 100)
```

```python
import matplotlib.pyplot as plt 
plt.plot(gdp_cap,life_exp)

# Definition of tick_val and tick_lab
tick_val = [5000,10000]
tick_lab = ['5k','10k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

plt.grid(True) 

plt.show()
```

```python
import numpy as np

plt.scatter(gdp_cap, life_exp, s = np.array(pop) ** 2, c = col, alpha = 0.8)
# Put the x-axis on a logarithmic scale
plt.xscale('log')

# Add axis labels
plt.xlabel('GDP per Capita [in USD]') 
plt.ylabel('Life Expectancy [in years]')

# Add title
plt.title('World Development in 2007')

plt.text(4550, 171, 'hi there')

plt.show()
```

```python
plt.hist(life_exp, bins=5)
#plt.show()
plt.hist(life_exp, bins=20)
plt.show()
```

```python
plt.hist(life_exp, bins=5)
plt.show()
plt.hist(life_exp, bins=20)
plt.show()
```

## loc and iloc

**loc** is label-based, which means that you have to specify rows and columns based on their row and column labels.

**iloc** is integer index based, so you have to specify rows and columns by their integer index.

```python
print(cars.loc[['c', 'b'], 'drives_right'].head(3))
```

```python
print(cars.iloc[1:5])
```

```python
print(cars.iloc[[1,5]])
```

 ## Logic, Control Flow and Filtering

```python
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

# my_house greater than 18.5 or smaller than 10
print(np.logical_or(my_house > 18.5,  my_house < 10)
)
# Both my_house and your_house smaller than 11
print( np.logical_and(my_house <11, your_house <11))
```
